```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,100,10)
y=x*x
```

```python
print("x=:",x)
print("y=:",y)
plt.plot(x,y)
```

```python
x=['A','B','C']
y=[50,70,90]
print("X=:",x)
print("Y=:",y)
plt.title("Line chart demo")
plt.xlabel('x_value')
plt.ylabel('Y_value')
plt.plot(x,color='green',marker="*",linestyle='dashed',mec='r',mfc='b',linewidth=3,markersize=3)
```

```python
categories=np.array(['A','B','C','D'])
values=np.array([3,8,1,10])
value2=np.array([15,26,21,31])
value3=np.array([45,16,20,9])
bar_width=0.35
plt.bar(categories,values,bar_width,label='Dataset1')
plt.bar(categories,value2,bar_width-0.30,label='Dataset2')
plt.bar(categories,value3,bar_width-0.3,label='Dataset3')
plt.xlabel("categories")
plt.ylabel('value')
plt.title('Grouped bar chart Examples')
plt.legend()
plt.show()
```

```python
explode=(0,0,0.1,0)
plt.pie(values,labels=categories,explode=explode,autopct="%1.1f%%")
plt.title('Exploded pie chart examples')
plt.show()
```

```python
Items=np.array(['Electronics','Clothing','Grocery','Household'])
Sell=np.array([5000,6000,4000,3500])
plt.pie(values,labels=Items,explode=explode,autopct="%1.1f%%")
plt.title("Kharche ")
plt.show()
```

```python
Items=np.array(['Electronics','Clothing','Grocery','Household'])
Sell=np.array([5000,6000,4000,3500])
plt.title("Every year kharche")
plt.xlabel("Items")
plt.ylabel("Sell")
bar_width=0.35
plt.bar(Items,Sell,bar_width)
plt.show()
```
