```python
x1=df['sepal_width']
y1=df['species']
plt.bar(x1,y1)
plt.title('Species and their width')
plt.xlabel('Sepal_width')
plt.ylabel('speices')
plt.show()
```

```python
categories=df['species']
sl=df['sepal_length']
sw=df['sepal_width']
pl=df['petal_length']
pw=df['petal_width']
bar_width=0.10
plt.bar(categories,sl,bar_width,label='Sepal_Length')
plt.bar(categories,sw,bar_width-0.14,label='Sepal_width')
plt.bar(categories,pl,bar_width-0.16,label='Petal_Length')
plt.bar(categories,pw,bar_width-0.18,label='petal_width')
plt.xlabel('Speices')
plt.ylabel('values')
plt.title('Iris data')
plt.legend()
plt.show()
```

```python
plt.pie(sl,labels=categories)
plt.title("values and species")
plt.show()
```

```python
import numpy as np
```

```python
categories1=np.array(['a','b','c','d','e'])
values=np.array([2,5,4,10,6])
explode=(0,0,0.1,0,0)
plt.pie(values,labels=categories1,explode=explode,autopct="%1.1f%%",startangle=180,counterclock=False)
plt.title("values and species")
plt.show()
```

```python
fig, ax=plt.subplots()
ax.pie(values,labels=categories1,radius=1,startangle=90,counterclock=False,autopct="%1.1f%%")
ax.pie([1],radius=0.4,colors='w')
ax.set_title("Donut chart")
plt.show()
```
