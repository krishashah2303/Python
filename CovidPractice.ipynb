```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("country_wise_latest.csv")
df
```

```python
df.plot(kind="scatter",x='Recovered',y='Deaths')
plt.grid()
plt.plot()
```

```python
plt.plot(df['New cases'],df['New deaths'])
plt.show
```

```python
df['WHO Region'].value_counts()
```

```python
sns.set(style="darkgrid")
sns.scatterplot(x='Active',y='Deaths',hue='Country/Region',data=df)
```

```python
categories=df['Recovered']
values=df['Confirmed']
fig, ax=plt.subplots()
ax.pie(values,labels=categories,radius=1,startangle=90,counterclock=False,)
ax.pie([1],radius=0.6,colors='w')
ax.set_title("Dount chart example")
plt.show()
```

```python
bar_width=25
plt.bar(df['Country/Region'],df['Deaths'],bar_width,label='Location')
plt.bar(df['Country/Region'],df['Recovered'],bar_width-20,label='Recovered')
plt.ylabel('Country/Region')
plt.xlabel('Deaths')
plt.legend()
plt.show()
```

```python
sns.relplot(x='Deaths',y='Recovered',hue='WHO Region',data=df)
```

```python
x=df['Confirmed']
y=df['Confirmed last week']
plt.plot(x,y)
```

```python
location=df['WHO Region']
deaths=df['Deaths']
plt.title('Deths in diffrent Regions')
plt.xlabel('location')
plt.ylabel('deaths')
bar_width=0.35
plt.pie(deaths,labels=location)
plt.show()
```
